```python
import scipy.integrate
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def SIR_model(y, t, beta, gamma):
    S, I, R = y
    
    dS_dt = -beta*S*I
    dI_dt = beta*S*I - gamma*I
    dR_dt = gamma*I
    
    
    return ([dS_dt, dI_dt, dR_dt])
```

```python
factor = 1
factorI = 10
pop = 51315383
recov_col = 2406377*factor+63422
activ_col = 54636*factorI
init_conds = np.array([pop-(activ_col+recov_col), activ_col, recov_col])/pop

gamma = 1/14
r0 =  1.6 #beta/gamma
beta = r0*gamma

t = np.linspace(0,500,10000)
sol = scipy.integrate.odeint(SIR_model, init_conds, t, args=(beta,gamma))

plt.plot(t, sol[:,0]*pop/1e6)
plt.plot(t, sol[:,1]*pop/1e6)
plt.plot(t, sol[:,2]*pop/1e6)
plt.show()

plt.plot(t, -np.gradient(sol[:,0])*pop)
plt.grid()
plt.show()
```
